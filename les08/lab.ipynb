```python
plt.figure(figsize=(8, 6))

scatter = plt.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    c=df["cluster"],
    cmap="tab10",
    s=50
)

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("DBSCAN clustering (Iris dataset)")
plt.colorbar(scatter, label="Cluster")
plt.grid(True)
plt.show()
```
