4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
print(df['Pclass'].value_counts())
print(pd.crosstab(df['Pclass'], df['Sex']))
```

5. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python
median_fare = df['Fare'].median()
std_fare = df['Fare'].std()
print(f"Медиана: {median_fare:.2f}")
print(f"Стандартное отклонение: {std_fare:.2f}")
```

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
young = df[df['Age'] < 30]['Survived'].mean()
old = df[df['Age'] > 60]['Survived'].mean()
print(f"Доля выживших младше 30: {young:.3f}")
print(f"Доля выживших старше 60: {old:.3f}")
print("Да, люди моложе 30 выживали чаще.")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
female_survival = df[df['Sex'] == 'female']['Survived'].mean()
male_survival = df[df['Sex'] == 'male']['Survived'].mean()
print(f"Доля выживших женщин: {female_survival:.3f}")
print(f"Доля выживших мужчин: {male_survival:.3f}")
print("Да, женщины выживали чаще.")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
df_male = df[df['Sex'] == 'male']['Name']

def get_first_name(name):
    if '(' in name:
        return name.split('(')[0].split(',')[1].strip().split()[0]
    else:
        return name.split(',')[1].strip().split()[0]

first_names = df_male.apply(get_first_name)
most_common = first_names.value_counts().idxmax()
print(f"Самое популярное имя: {most_common}")
```



9. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

    В среднем мужчины 1-го класса старше 40 лет
    В среднем женщины 1-го класса старше 40 лет
    Мужчины всех классов в среднем старше женщин того же класса
    В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса



```python
avg_age = df.groupby(['Pclass', 'Sex'])['Age'].mean()
print(avg_age)
print("""
Верные утверждения:
- В среднем мужчины 1-го класса старше 40 лет: Да (42.0 > 40)
- В среднем женщины 1-го класса старше 40 лет: Нет (34.9 < 40)
- Мужчины всех классов в среднем старше женщин того же класса: Да
- В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса: Да
""")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
survived_age = df[df['Survived'] == 1]['Age'].mean()
died_age = df[df['Survived'] == 0]['Age'].mean()
print(f"Средний возраст выживших: {survived_age:.1f}")
print(f"Средний возраст погибших: {died_age:.1f}")
print("Да, средний возраст погибших выше.")
```
