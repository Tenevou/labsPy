1. Создать массив 10x10 со случайными значениями, найти минимум и максимум (низкая)

````python
arr = np.random.randint(0, 100, (10, 10))

print("Массив:")
print(arr)
print("Минимум:", arr.min())
print("Максимум:", arr.max())
````

2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

````python
import numpy as np

chessboard = np.zeros((8, 8))
chessboard[1::2, ::2] = 1
chessboard[::2, 1::2] = 1

print("Шахматная матрица:")
print(chessboard)
````

3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

````python
z = np.arange(11)
mask = (z >= 3) & (z <= 8)
z[mask] *= -1

print("Изменённый массив:", z)
````

4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

````python
z = np.random.rand(5, 10)
z_centered = z - z.mean(axis=1, keepdims=True)

print("Исходная матрица:\n", z)
print("\nЦентрированная по строкам:\n", z_centered)
````

5. Поменяйте 1 и 2 строки матрицы

````python
z = np.arange(25).reshape(5, 5)
z[[1, 2]] = z[[2, 1]]

print("Изменённая матрица:\n", z)
````

6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
z = np.random.random((10, 2))
x, y = z[:, 0], z[:, 1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

polar = np.column_stack((r, theta))

print("Декартовы координаты:\n", z)
print("\nПолярные координаты (r, theta):\n", polar)
````

7. Заменить максимальный элемент массива на ноль

````python
z = np.random.random(10)
max_idx = z.argmax()
z[max_idx] = 0

print("Изменённый массив:", z)
````

8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

````python
z = np.random.random(10)
indices = np.array([1, 3, 5, 3])

np.add.at(z, indices, 1)

print("Изменённый массив:", z)
````

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
vec = np.array([1, 2, 3, 4, 5])
new_vec = np.empty(len(vec) * 4 - 3, dtype=int)
new_vec[0::4] = vec
new_vec[1:4] = 0
new_vec[5:8] = 0
new_vec[9:12] = 0
new_vec[13:16] = 0

print("Новый вектор:", new_vec)
````

10. Найти ближайшее к заданному значению число в заданном массиве

````python
z = np.random.random(10)
target = 3.14

closest_idx = np.abs(z - target).argmin()
closest_value = z[closest_idx]

print("Массив:", z)
print("Целевое значение:", target)
print("Ближайшее значение:", closest_value)
````
